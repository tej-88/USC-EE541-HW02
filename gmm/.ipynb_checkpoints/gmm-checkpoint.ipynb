```python
#Importing Libraries
import re
import csv
import numpy as np
from math import log
from numpy.linalg import norm, det, pinv
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
#Constants
MICKEY_FILENAME = './mickey.csv'
K = 3
KMEANS_LABELS_FILENAME = './kmeans_label.csv'
```

```python
def get_data(filename):
    data = []
    labels_dict = {'head' : 2, 'ear_left' : 1, 'ear_right' : 3}

    with open(filename, 'r') as f:
        reader = csv.reader(f)

        for row in reader:
            if not re.search('^#', row[0]):
                actual_row = row[0].split()
                actual_row[0] = float(actual_row[0])
                actual_row[1] = float(actual_row[1])
                actual_row[-1] = labels_dict[actual_row[-1].lower()]
                data.append(actual_row)

    data = np.array(data)
    X = data[:, :-1]
    y = data[:, -1]
    
    return (X, y)
```

```python
def get_label_colors(y):
    colors_dict = {2 : 'b', 1 : 'r', 3 : 'g'}
    y_colors = []
    
    for i in y:
        y_colors.append(colors_dict[int(i)])
        
    return y_colors
```

```python
class GMM():
    
    def __init__(self):
        self.D = 0
        self.X = None
        self.mean = None
        self.sigma = None
        self.weight = None
        
    def fit(self, X, y):
        pass
    
    
    def get_gaussian_pdf(self, x, k):
        mean = self.mean[k, :]
        sigma = self.sigma[k, :]
        mean_diff = x - mean
        sigma_inv = pinv(sigma)
        exponent = -(0.5) * np.dot(mean_diff.T, np.dot(sigma_inv, mean_diff))
        numerator = np.exp(exponent)
        sigma_det = det(sigma)
        denominator = np.sqrt(((np.pi * 2) ** self.D) * sigma_det)
        return (numerator / denominator)
    
    
    def get_hidden_variable(self, x, k):
        numerator = self.weight[k] * get_gaussian_pdf(x, k)
        denominator = 0
        for i in range(1, self.K + 1):
            denominator += self.weight[i] * get_gaussian_pdf(x, i)
        return (numerator / denominator)
    
    
    def update_mean_k(self, k):
        numerator = None
        denominator = 0
        hidden_variable = 0
        for i in self.X:
            hidden_variable = get_hidden_variable(i, k)
            numerator += hidden_variable * i
            denominator += hidden_variable
        self.mean[k, :] = (numerator / denominator)
        return
    
    
    def update_sigma_k(self, k):
        numerator = None
        denominator = 0
        hidden_variable = 0
        mean = self.mean[k, :]
        for i in self.X:
            mean_diff = i - mean
            hidden_variable = get_hidden_variable(i, k)
            numerator += hidden_variable * np.dot(mean_diff, mean_diff.T)
            denominator += hidden_variable
        self.sigma[k, :] = (numerator / denominator)
        return
    
    def update_weight_k(self, k):
        weight = 0
        for i in self.X:
            weight += get_hidden_variable(i, k)
        self.weight[k] = (weight / len(self.X))
        return
    
    def one_hot_encoding(self, y):
        y_encoding = np.zeros((len(y), self.D))

        for i in range(len(y)):
            y_encoding[i, y[i] - 1] = 1.0

        return y_encoding
```

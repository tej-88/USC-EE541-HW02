```python
#Importing Libraries
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.interpolate import CubicSpline
```

```python
#Constants
RAMAN_DATASET_FILE = './raman.txt'
NUM_REGIONS_OF_INTEREST = 4
```

```python
#Read the Data
raman_data = np.loadtxt(RAMAN_DATASET_FILE)
wavenumber = raman_data[:, 0]
intensity = raman_data[:, 1]
```

```python
#Find Spectral Peaks
peak_idx = find_peaks(intensity, height=3000, distance = 300)[0][:NUM_REGIONS_OF_INTEREST]
```

```python
#Figure
plt.figure()
plt.plot(wavenumber, intensity)
plt.scatter(wavenumber[peak_idx], intensity[peak_idx], c='#000000')

plt.xlabel('Wavenumber (in cm-1)')
plt.ylabel('Photon Intensity')
plt.title('Raman Spectroscopy')

plt.legend(['Raman Spectroscopy', 'Spectral Peak'])
plt.show()
```

```python
good_widths = np.empty((NUM_REGIONS_OF_INTEREST))
```

```python
def get_good_widths(peak_idx):
    widths = np.arange(50, 0, -1)
    threshold = intensity[peak_idx] / 2
    for width in widths:
        if intensity[peak_idx] - intensity[peak_idx - width] < threshold :
            return width
```

```python
for i in range(len(peak_idx)):
    good_widths[i] = get_good_widths(peak_idx[i])
```

```python
print('#' * 50)
for j in range(NUM_REGIONS_OF_INTEREST):
    neighborhood = np.arange(peak_idx[j] - good_widths[j], peak_idx[j] + good_widths[j] + 1, dtype=np.int32)
    intensity_interpolate = CubicSpline(wavenumber[neighborhood], intensity[neighborhood])
    wavenumber_list = np.linspace(wavenumber[neighborhood[0]], wavenumber[neighborhood[-1]], num=100)
    derivative_list = np.array(list(map(lambda x: (intensity_interpolate(x + 0.01) - intensity_interpolate(x - 0.01)) / (0.02), wavenumber_list)))
    zero_crossing = list(filter(lambda x: x < 0, derivative_list))[0]
    peak_wavenumber = wavenumber_list[derivative_list == zero_crossing][0]
    
    plt.figure()
    plt.scatter(wavenumber[neighborhood], intensity[neighborhood], c='#000000')
    plt.plot(wavenumber[neighborhood], intensity_interpolate(wavenumber[neighborhood]))
    plt.scatter(peak_wavenumber, intensity_interpolate(peak_wavenumber), c='#FF0000', s=100)
    plt.scatter(peak_wavenumber, 0, c='#FF00FF', s=100)
    
    print('Peak Wavenumber(in cm-1): ')
    sys.stdout.write(f'{peak_wavenumber}')
    
    plt.xlabel('Wavenumber (in cm-1)')
    plt.ylabel('Photon Intensity')
    plt.title(f'Raman Spectroscopy around spectral peak {wavenumber[peak_idx[j]]} cm-1')
    plt.legend(['Raman Spectroscopy', 'Interpolation', 'Spectral Peak', 'Peak Wavenumber'])

    plt.show()
    print('#' * 50)
```
